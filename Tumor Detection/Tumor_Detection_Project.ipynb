```python
df['diagnosis']=df['diagnosis'].map({'M' : 1, 'B' :10}) # Makes it numeric so ML models can use it.
```

```python
df['diagnosis'].unique() # Verify encoding.
```

```python
# Split Features and Target
x = df.drop('diagnosis', axis=1)
x.head()
```

```python
y= df['diagnosis']
y
```

```python
# Training–Test Split
# Machine Learning libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Shows number of rows/columns in train & test sets.
df.shape
X_train.shape
X_test.shape
y_train.shape
y_test.shape
```

```python
# Standardization
# Makes all features on similar scale (mean=0, SD=1).
# Only fit training data (important to avoid leakage).
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
# Import ML model and accuracy metric.
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Apply Random Forest Classifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

accuracy
```
